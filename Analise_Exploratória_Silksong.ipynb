```python
import zipfile

# Descompactar o arquivo
with zipfile.ZipFile("classical-music-midi.zip", "r") as zip_ref:
    zip_ref.extractall("classical-music-midi_data")
```

```python
!pip install pretty_midi
import pretty_midi

midi_data = pretty_midi.PrettyMIDI("classical-music-midi_data/bach/bach_846.mid")

for instrument in midi_data.instruments:
    # lista os instrumentos contidos no arquivo, no caso, apenas piano (dividido em mão esquerda e direita) e fugas
    print(f"Instrument: {instrument.name}")
    for note in instrument.notes:
        # printa o tom da nota, quando ela comeca, quando termina, e o quao rapida ela é tocada
        print(f"Note: {note.pitch}, Start: {note.start}, End: {note.end}, Velocity: {note.velocity}")
```

```python
# estima o tempo da musica
print(midi_data.estimate_tempo())
```
